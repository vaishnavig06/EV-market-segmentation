```python
import pandas as pd 
import numpy as np                      
import seaborn as sns                  
import matplotlib.pyplot as plt  

data = pd.read_csv('pollution data.csv')
data
```

```python
data.head()
data.info()
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x='state', y='AQI-US', data=data, palette='viridis')
plt.xticks(rotation=90)
plt.title('Air Quality Index (AQI-US) by State')
plt.xlabel('State')
plt.ylabel('AQI-US')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='status', y='PM2.5', data=data, palette='Set2')
plt.title('Distribution of PM2.5 by Air Quality Status')
plt.xlabel('Air Quality Status')
plt.ylabel('PM2.5')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='status', y='PM10', data=data, palette='Set3')
plt.title('Distribution of PM10 by Air Quality Status')
plt.xlabel('Air Quality Status')
plt.ylabel('PM10')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temp', y='Humid', hue='AQI-US', data=data, palette='coolwarm', size='AQI-US', sizes=(20, 200))
plt.title('Temperature vs Humidity Colored by AQI-US')
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity (%)')
plt.show()
```

```python
data1 = pd.read_csv('states_data_car.csv')
data1
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
features = data1[['subsidy', 'road tax', 'petrol', 'diesel']]
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
kmeans = KMeans(n_clusters=3, random_state=42)
data1['Cluster'] = kmeans.fit_predict(features_scaled)
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x='state', y='petrol', data=data1, palette='Blues_d')
plt.xticks(rotation=90)
plt.title('Petrol Prices by State')
plt.xlabel('State')
plt.ylabel('Petrol Price')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x='state', y='diesel', data=data1, palette='Oranges_d')
plt.xticks(rotation=90)
plt.title('Diesel Prices by State')
plt.xlabel('State')
plt.ylabel('Diesel Price')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Cluster', y='subsidy', data=data1, palette='Set2')
plt.title('Distribution of Subsidy by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Subsidy')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Cluster', y='road tax', data=data1, palette='Set3')
plt.title('Distribution of Road Tax by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Road Tax')
plt.show()
```

```python

# Select relevant features for clustering (assuming 'Petrol_Price' and 'Diesel_Price' are the correct columns)
features = data1[['petrol', 'diesel']]

# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
data1['Cluster'] = kmeans.fit_predict(features_scaled)

# Create a scatter plot of Petrol vs. Diesel Prices, color-coded by cluster
plt.figure(figsize=(10, 6))
sns.scatterplot(x='petrol', y='diesel', hue='Cluster', data=data1, palette='Set2')
plt.title('Scatter Plot of Petrol vs. Diesel Prices by Cluster')
plt.xlabel('petrol')
plt.ylabel('diesel')
plt.show()
```
